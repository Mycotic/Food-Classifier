```python
vege_codes = [1,2,8,9,11,12,16,20]
meat_codes = [5,7,10,13,15,17]
vege_cats_brands = [
    "Cheese",
    "Ice Cream & Frozen Yogurt",
    "Cookies & Biscuits",
    "Breads & Buns",
    "Pickles, Olives, Peppers & Relishes",
    "Fruit & Vegetable Juice, Nectars & Fruit Drinks",
    "Cereal",
    "Chocolate",
    "Cakes, Cupcakes, Snack Cakes",
    "Soda",
    "Yogurt",
    "Ketchup, Mustard, BBQ & Cheese Sauce",
    "Pasta by Shape & Type",
    "Water",
    "Canned Vegetables",
    "Pre-Packaged Fruit & Vegetables",
    "Frozen Vegetables",
    "Milk",
    "Canned Fruit",
    "Powdered Drinks",
    'Nut & Seed Butters',
    "Energy, Protein & Muscle Recovery Drinks",
    "French Fries, Potatoes & Onion Rings"
]
meat_cats_brands = [
    "Pepperoni, Salami & Cold Cuts",
    "Sausages, Hotdogs & Brats",
    "Frozen Fish & Seafood",
    "Other Meats",
    "Bacon, Sausages & Ribs",
    "Chili & Stew",
    "Canned Tuna",
    "Fish & Seafood",
    "Meat/Poultry/Other Animals Prepared/Processed"
]
```

```python
def meat_me(row):
    # returns 1 if meat 0 if vegetarian, and -1 if not sure or too small a category
    if row["branded_food_category"] in meat_cats_brands or row.loc["food_category_id"] in meat_codes:
        return 1
    elif row["branded_food_category"] in vege_cats_brands or row.loc["food_category_id"] in vege_codes:
        return 0
    return -1
```

```python
#food_df_merge["is_meat"] = 
food_df_merge["is_meat"]=food_df_merge.apply(meat_me,axis=1)
food_df_merge = food_df_merge[food_df_merge["is_meat"]!=-1]
food_df_merge.shape
```

```python
# we should remove foods that have more than 100 grams
# of a nutrient per 100grams of that food (it's ok to remove from
# test set here as well bc its data that is strictly impossible)
# some samples are measured in other units than grams so lets
# look at just gram features to start with
gram_nutrients = nutrient_names[nutrient_names["unit_name"]=="G"]["name"]
final_food_df[gram_nutrients].max()
```

```python
# yea looks like we have some erroneous data... so I'll drop
# a row if it has more than 100g
less_100g = final_food_df[gram_nutrients].apply(lambda x: max(x)<=100,axis=1)
```

```python
# my suspicion is that most of the errors come from using milligrams instead of
# grams - i dont think just dividing them by 1k is actually safe though so just
# going to drop them.
# using if to stop it from breaking second time through
if len(food_df_merge)==len(final_food_df):
    print("dropping outliers")
    final_food_df=final_food_df[less_100g]
    target = final_food_df["is_meat"]
    final_food_df = final_food_df.drop("is_meat",axis=1)
```
